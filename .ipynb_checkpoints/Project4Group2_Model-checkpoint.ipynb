```python
# Import required libraries and dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data into a Pandas DataFrame
df_pet_adoption = pd.read_csv(
    "Output/pet_adoption_renamed.csv", index_col="Pet_ID")

# Display sample data
df_pet_adoption.head(10)
```

```python
df_pet_adoption.columns.values.tolist()
```

```python
# imports
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# Review the info
df_pet_adoption.info()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
df_pet_adoption_dummies = pd.get_dummies(df_pet_adoption, columns=[
    'Pet_Type', 'Breed', 'Color', 'Size'
])
# Display the transformed data
df_pet_adoption_dummies
```

```python
df_pet_adoption_dummies.columns.values.tolist()
```

# Create a Logistic Regression Model 

```python
# Separate the data into labels and features

# Separate the y variable, the labels
y = df_pet_adoption_dummies["Adoption_Likelihood"]

# Separate the X variable, the features
X = df_pet_adoption_dummies.drop(columns=["Adoption_Likelihood"])
```

```python
# Split the data using train_test_split
# Assign a random_state of 1 to the function
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
X_train.shape
```

```python
# Import the LogisticRegression module from SKLearn
from sklearn.linear_model import LogisticRegression

# Ensure y_train is a 1D array
y_train = y_train.ravel()

# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model

logistic_regression_model = LogisticRegression(solver='lbfgs',max_iter=1000, random_state=1)

# Fit the model using training data
lr_model = logistic_regression_model.fit(X_train, y_train)
```

```python
#Score the model using the test data

print(f"Training Data Score: {lr_model.score(X_train, y_train)}")
print(f"Testing Data Score: {lr_model.score(X_test, y_test)}")
```

```python
# Make a prediction using the testing data
#Generate testing and training predictions

training_predictions = lr_model.predict(X_train)
testing_predictions = logistic_regression_model.predict(X_test)
```

```python
# Predict outcomes for test data set and show in a dataframe
results_df = pd.DataFrame({"Prediction": testing_predictions, "Actual":y_test})
results_df.reset_index(drop=True, inplace=True)
results_df
```

```python
from sklearn.metrics import accuracy_score
# Display the accuracy score for the test dataset.
accuracy_score(y_test, testing_predictions)
```

```python
# Generate a confusion matrix for the model
confusion_matrix(y_test, testing_predictions)
```

```python
# Print the classification report for the model
testing_report = classification_report(y_test, testing_predictions)

# Print the testing classification report
print(testing_report)
```

## Matrix and score on training data for comparison

```python
# Create and save the confusion matrix for the training data
training_matrix = confusion_matrix(y_train, training_predictions)

# Print the confusion matrix for the training data
print(training_matrix)
```

```python
# Create and save the training classification report
training_report = classification_report(y_train, training_predictions)

# Print the training classification report
print(training_report)
```

# Predicting Adoption with Random Forest Model

```python
# Define features set
X_rf = df_pet_adoption_dummies.copy()
X_rf.drop("Adoption_Likelihood", axis=1, inplace=True)
X_rf.head()
```

```python
# Define target vector
y_rf = df_pet_adoption_dummies["Adoption_Likelihood"].ravel()

y_rf[:5]
```

```python
# Splitting into Train and Test sets
X_rf_train, X_rf_test, y_rf_train, y_rf_test = train_test_split(X_rf, y_rf, random_state=78)
```

```python
# Create the StandardScaler instance
scaler = StandardScaler()
```

```python
# Fit the Standard Scaler with the training data
X_scaler = scaler.fit(X_rf_train)
```

```python
# Scale the training data
X_train_scaled = X_scaler.transform(X_rf_train)
X_test_scaled = X_scaler.transform(X_rf_test)
```

## Fitting the Random Forest Model


```python
# Create a random forest classifier
rf_model = RandomForestClassifier(n_estimators=500, random_state=78)
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_rf_train)
```

## Making Predictions Using the Random Forest Model

```python
# Making predictions using the testing data
predictions = rf_model.predict(X_test_scaled)
```

## Model Evaluation

```python
# Calculating the confusion matrix
cm = confusion_matrix(y_rf_test, predictions)
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]
)

# Calculating the accuracy score
acc_score = accuracy_score(y_rf_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_rf_test, predictions))
```

## Feature Importance

```python
# Get the feature importance array

importances = rf_model.feature_importances_
feature_names = X.columns

# Create a DataFrame for feature importances and sort it
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# List the top 10 most important features
feature_importance_df.head(10)
```

```python
# Visualize the features by importance
importances_df = pd.DataFrame(sorted(zip(rf_model.feature_importances_, X.columns), reverse=True))
importances_df.set_index(importances_df[1], inplace=True)
importances_df.drop(columns=1, inplace=True)
importances_df.rename(columns={0: 'Feature Importances'}, inplace=True)
importances_sorted = importances_df.sort_values(by='Feature Importances')
importances_sorted.plot(kind='barh', color='lightgreen', title= 'Features Importances', legend=False)
plt.tight_layout()
plt.savefig("output_dir/features_by_importance.png")
```

## Optimizing Random Forest Model, by dropping insignificant Features (Color, Previous Owner)


```python
# show the original df

df_pet_adoption.head()
```

```python
# drop unnecessary columns

df_pet_adoption_reduced = df_pet_adoption.copy()
df_pet_adoption_reduced.drop(columns=["Color", "Previous_Owner"], inplace=True)
df_pet_adoption_reduced.head()
```

```python
# Review the info
df_pet_adoption_reduced.info()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
df_pet_adoption_reduced_dummies = pd.get_dummies(df_pet_adoption_reduced, columns=[
    'Pet_Type', 'Breed','Size'
])
# Display the transformed data
df_pet_adoption_reduced_dummies
```

```python
# Define features set
X_rf_reduced = df_pet_adoption_reduced_dummies.copy()
X_rf_reduced.drop("Adoption_Likelihood", axis=1, inplace=True)
X_rf_reduced.head()
```

```python
# Define target vector
y_rf_reduced = df_pet_adoption_reduced_dummies["Adoption_Likelihood"].ravel()

y_rf_reduced[:5]
```

```python
# Separate the data into labels and features

# Separate the y variable, the labels
y_rf_reduced = df_pet_adoption_reduced_dummies["Adoption_Likelihood"]

# Separate the X variable, the features
X_rf_reduced = df_pet_adoption_reduced_dummies.drop(columns=["Adoption_Likelihood"])
```

```python
# Splitting into Train and Test sets
X_rf_reduced_train, X_rf_reduced_test, y_rf_reduced_train, y_rf_reduced_test = train_test_split(X_rf_reduced, y_rf_reduced, random_state=78)
```

```python
# Create the StandardScaler instance
scaler_reduced = StandardScaler()
```

```python
# Fit the Standard Scaler with the training data
X_scaler_reduced = scaler_reduced.fit(X_rf_reduced_train)
```

```python
# Scale the training data
X_train_reduced_scaled = X_scaler_reduced.transform(X_rf_reduced_train)
X_test_reduced_scaled = X_scaler_reduced.transform(X_rf_reduced_test)
```

```python
# Create a random forest classifier
rf_model_reduced = RandomForestClassifier(n_estimators=500, random_state=78)
```

```python
# Fitting the model
rf_model_reduced = rf_model_reduced.fit(X_train_reduced_scaled, y_rf_reduced_train)
```

```python
# Making predictions using the testing data
predictions_reduced = rf_model_reduced.predict(X_test_reduced_scaled)
```

```python
# Calculating the confusion matrix
cm_r = confusion_matrix(y_rf_reduced_test, predictions_reduced)
cm_r_df = pd.DataFrame(
    cm_r, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]
)

# Calculating the accuracy score
acc_score_r = accuracy_score(y_rf_reduced_test, predictions_reduced)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_r_df)
print(f"Accuracy Score : {acc_score_r}")
print("Classification Report")
print(classification_report(y_rf_reduced_test, predictions_reduced))
```
